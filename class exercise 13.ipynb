Objective: Understand the K-means clustering by applying it to the iris dataset and observe how it clusters the data based on flower features.

## 1. Inport libraries and set up the environment:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

## 2. Load the dataset:

```python
iris = datasets.load_iris()
df = pd.DataFrame(data = iris.data, columns=iris.feature_names)
```

## 3. Data Exploration:

```python
df.head()
df.describe()
```

## 4. Data Preprocessing: (Scale the data in K-means using StandardScaler!)

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

## 5. Apply K-means clustering (start with k = 3

```python
kmeans = KMeans(n_clusters = 3, random_state = 42)
clusters = kmeans.fit_predict(df_scaled)
df['cluster'] = clusters
```

## 6. Visualize clusters

```python
plt.scatter(df_scaled[clusters == 0,0], df_scaled[clusters == 0,1], label = 'Cluster 1')
plt.scatter(df_scaled[clusters == 1,0], df_scaled[clusters == 1,1], label = 'Cluster 2')
plt.scatter(df_scaled[clusters == 2,0], df_scaled[clusters == 2,1], label = 'Cluster 3')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], s = 300, c = 'red',label = 'Centroid')
plt.xlabel('Sepal Length')
plt.xlabel('Sepal Width')
plt.legend()
plt.show()
```

## 7. Analyze results:

### Apply K-means clustering (start with k = 6

```python
kmeans = KMeans(n_clusters = 6, random_state = 42)
clusters = kmeans.fit_predict(df_scaled)
df['cluster'] = clusters
```

```python
plt.scatter(df_scaled[clusters == 0,0], df_scaled[clusters == 0,1], label = 'Cluster 1')
plt.scatter(df_scaled[clusters == 1,0], df_scaled[clusters == 1,1], label = 'Cluster 2')
plt.scatter(df_scaled[clusters == 2,0], df_scaled[clusters == 2,1], label = 'Cluster 3')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], s = 300, c = 'red',label = 'Centroid')
plt.xlabel('Sepal Length')
plt.xlabel('Sepal Width')
plt.legend()
plt.show()
```
