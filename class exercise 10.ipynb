```python
#Import necessary libraries
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

```

#### Using Gini

```python
#Load dataset
iris = load_iris()
x = iris.data
y = iris.target
```

```python
#Split dataset into training set and test set, 80% training and 20% testing
x_train,x_test, y_train,y_test = train_test_split(x, y, test_size = 0.2)
```

```python
#Create a Decition Tree Classifier(using the Gini criterion)
clf = DecisionTreeClassifier(criterion = "gini")
```

```python
#Train Decisiom Tree Clasiffier
clf = clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
print("Accuracy: ", accuracy_score(y_test, y_pred))
```

```python
plt.figure(figsize=(20, 15))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names.tolist(), rounded=True)
plt.show()
```

#### Using entropy

```python
#Load dataset
iris = load_iris()
x = iris.data
y = iris.target


#Split dataset into training set and test set, 80% training and 20% testing
x_train,x_test, y_train,y_test = train_test_split(x, y, test_size = 0.2)


#Create a Decition Tree Classifier(using the Gini criterion)
clf = DecisionTreeClassifier(criterion = "entropy")


#Train Decisiom Tree Clasiffier
clf = clf.fit(x_train,y_train)

y_pred = clf.predict(x_test)

print("Accuracy: ", accuracy_score(y_test, y_pred))

plt.figure(figsize=(20, 15))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names.tolist(), rounded=True)
plt.show()
```
